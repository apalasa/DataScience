```
#1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
sns.set()

#download data
! if [ ! -e orings.csv ]; \
then wget https://raw.githubusercontent.com/bcaffo/ds4bme_intro/master/data/orings.csv; \
fi;


#a  
## Read in the data
dat = pd.read_csv('https://raw.githubusercontent.com/bcaffo/ds4bme_intro/master/data/orings.csv')

total = np.array(dat['Total'])
temp = np.array(dat['Temperature'])

for i in range(len(total)):
  if total[i] != 0:
    total[i]=1
    
sns.scatterplot(temp, total)
plt.xlabel("Temperature (F)")
plt.ylabel("Oring Failure (total>0)")
plt.title("Oring Failure over Temperature")

#b
temp = temp.reshape(-1,1)
lr = lm.LogisticRegression(fit_intercept=True, solver = 'lbfgs', penalty = 'none')
fit = lr.fit(temp, total)

#c
beta0, beta1 = [fit.intercept_[0], fit.coef_[0][0]]
[beta0, beta1]
n = 10000
xplot = np.linspace(50, 90, n)
eta = beta0 + beta1 * xplot
p = 1 / (1 + np.exp(-eta))
sns.lineplot(xplot, p)
plt.legend(['Logistic Regression','Collected Data'])

print ('Beta_0 = %f, Beta_1 = %f' %(beta0, beta1))

#d x= 74
eta_2 = beta0 + beta1 * 74
p = 1 / (1 + np.exp(-eta_2))
print ('The estimated probability of failure from my model is %f' %p)
```
